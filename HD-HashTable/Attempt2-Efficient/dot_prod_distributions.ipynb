```python
import matplotlib.pyplot as plt
```

```python
# Dot products for queries when D=10000
f = open('dot_prods_D_10000.csv', 'r')

dot_prod_pos = []
dot_prod_neg = []

data = f.readline()
while data != '':
  data = data.split(',')
  dot_prod = int(data[0])
  if data[1].startswith('T'):
    dot_prod_pos.append(dot_prod)
  elif data[1].startswith('F'):
    dot_prod_neg.append(dot_prod)
  data = f.readline()
f.close()
```

```python
plt.hist([dot_prod_pos, dot_prod_neg],
    bins=30,
    color=['blue', 'red'],
    edgecolor='black',
    label=['In Hash Table', 'Not In Hash Table'])

plt.xlabel('Dot Product')
plt.ylabel('Frequency')
plt.title('Histogram of Dot Products for K-mers for D=10000')
plt.legend()

plt.show()
```
