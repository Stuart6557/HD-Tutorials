```python
import matplotlib.pyplot as plt
```

```python
reads = [30, 40, 50, 100, 150, 200]
acc = [88.16, 85.75, 85.75, 79.50, 74.57, 72.50]
tpr = [87.78, 84.00, 85.50, 87.00, 91.33, 88.50]
tnr = [88.50, 87.50, 86.00, 72.00, 62.00, 56.50]

# scatter plots
plt.scatter(reads, acc, label='Overall Accuracy', color='r')
plt.scatter(reads, tpr, label='True Positive Rate', color='g')
plt.scatter(reads, tnr, label='True Negative Rate', color='b')

# line segments between points
plt.plot(reads, acc, color='r', linestyle='-', marker='o')
plt.plot(reads, tpr, color='g', linestyle='-', marker='o')
plt.plot(reads, tnr, color='b', linestyle='-', marker='o')

plt.xlabel('Number of reads', fontsize=20)
plt.ylabel('Accuracy (%)', fontsize=20)
plt.title('Query Accuracy for Varying Dataset Sizes', fontsize=20)
plt.legend(fontsize=20)

plt.rc('xtick', labelsize =20) 
plt.rc('ytick', labelsize=20)

plt.show()
```

```python
reads = [30, 40, 50, 100, 150, 200]
train_latency = [1337.63, 1801.06, 2201.16, 4317.61, 6222.66, 8413.00]
query_latency = [1.17, 1.27, 1.43, 1.96, 2.53, 2.89]

# scatter plots
plt.scatter(reads, train_latency, label='Training Latency', color='b')

# line segments between points
plt.plot(reads, train_latency, color='b', linestyle='-', marker='o')

# Add labels for each point
for i, qlat in enumerate(query_latency):
    plt.text(reads[i], train_latency[i], f'{qlat}s', fontsize=18, ha='right')

plt.xlabel('Number of reads', fontsize=20)
plt.ylabel('Latency (s)', fontsize=20)
plt.title('Latency for Varying Dataset Sizes', fontsize=20)
plt.legend(fontsize=20)

plt.rc('xtick', labelsize =20) 
plt.rc('ytick', labelsize=20)

plt.show()
```
