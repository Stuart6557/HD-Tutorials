## Experiments with First 2 Bacillus Reads

```python
import matplotlib.pyplot as plt
```

##### Dot products for queries when each hypervector contains the encoding for 1000 501-mers

```python
f = open('dot_prods.csv', 'r')

dot_prod_pos = []
dot_prod_neg = []

data = f.readline()
while data != '':
  data = data.split(',')
  dot_prod = int(data[0])
  if data[1].startswith('T'):
    dot_prod_pos.append(dot_prod)
  elif data[1].startswith('F'):
    dot_prod_neg.append(dot_prod)
  data = f.readline()
f.close()
```

```python
plt.hist([dot_prod_pos, dot_prod_neg],
    bins=30,
    color=['blue', 'red'],
    edgecolor='black',
    label=['In Hash Table', 'Not In Hash Table'])

plt.xlabel('Max Dot Product')
plt.ylabel('Frequency')
plt.title('Max Dot Products When Each HV Contains The Encoding For 1000 501-mers')
plt.legend()

plt.show()
```
